## Parte 1 - Data ETL, Data Wrangling e Data Exploration

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# upload e leitura dos dados

files = os.listdir('/content/')
df = dict()

for file in files:
  if file.endswith('.csv'):
    df[file] = dict()
    interval = str(file).split('-')[1]
    df[file]['signals'] = pd.read_csv(f'/content/{file}')
    df[file]['interval'] = interval

print(f'Leitura realizada dos arquivos: \n{", ".join(df.keys())}')

```

### 1. Apresentar os dados contidos nos arquivos no domínio do tempo.

```python
for file, content in df.items():
    signals = content['signals']
    interval = int(content['interval'])

    num_rows = len(signals)

    time = np.linspace(0, interval, num_rows)

    fig, axs = plt.subplots(3, 1, figsize=(10, 12))

    axs[0].plot(time, signals['x'], color='r', label='x')
    axs[0].set_title(f'{file} - eixo X  (Interval: {interval})')

    axs[1].plot(time, signals['y'], color='g', label='y')
    axs[1].set_title(f'{file} - eixo Y (Interval: {interval})')

    axs[2].plot(time, signals['z'], color='b', label='z')
    axs[2].set_title(f'{file} - eixo Z (Interval: {interval})')

    for ax in axs:
        ax.set_xlabel('Tempo [ms]')
        ax.set_ylabel('Aceleração [g]')
        ax.grid(True)
        ax.legend()

    plt.tight_layout()
    plt.show()
```
